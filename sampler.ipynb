## Poisson example

```python
Yn_poisson = init.poisson_data()
```

```python
Sn, F1_mcmc, F_mcmc, Theta_mcmc, P = sampler(Yn_poisson, model="poisson", m=1, cond=cond_opt)
```

```python
plt.subplot(2, 2, 1)
plt.bar(np.arange(len(Yn_poisson)), Yn_poisson)
plt.xlabel("Time")
plt.ylabel("yt")

plt.subplot(2, 2, 2)
density = gaussian_kde(Theta_mcmc[0, 1000:6000])
xs = np.linspace(2,5,200)
plt.plot(xs, density(xs))
plt.xlabel("lambda1")
plt.ylabel("pdf")

plt.subplot(2, 2, 3)
density = gaussian_kde(Theta_mcmc[1, 1000:6000])
xs = np.linspace(0.5,2,200)
plt.plot(xs, density(xs))
plt.xlabel("lambda2")
plt.ylabel("pdf")

plt.subplot(2, 2, 4)
plt.plot(F1_mcmc[:,:,1000:6000].mean(axis=-1));
plt.legend(['st=1', 'st=2'])
plt.xlabel("Time")
plt.ylabel("Pr(st|Yn)")

plt.tight_layout()
```
